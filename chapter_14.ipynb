```python
images = np.array([emma1, emma2])
batch_size, height, width, channels = images.shape
```

```python
filters = np.zeros(shape=(3, 3, channels, 2), dtype=np.float32)
filters[:, 2, :, 0] = 1 # vertical line
filters[:, 0, :, 0] = -1
filters[1, :, :, 0] *= 2
filters[2, :, :, 1] = 1 # horizontal line
filters[0, :, :, 1] = -1
filters[:, 1, :, 1] *= 2
```

```python
outputs = tf.nn.conv2d(images, filters, strides=1, padding='SAME')
```

```python
plt.imshow(outputs[1, :, :, 0], cmap='gray') 
plt.show()
```
